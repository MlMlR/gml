```python
import numpy as np
import matplotlib.pyplot as plt

def generate_dataset(n_points, beta_0, beta_1, sigma = 1, mu = 0):
    x = np.random.normal(mu, 1, n_points)
    eps = np.random.normal(0, sigma, n_points)
    y = beta_0 + beta_1 * x + eps
    
    return x, y

x, y = generate_dataset(n_points=100, beta_0=0, beta_1=1.5, sigma=1, mu=0)

a, b = generate_dataset(n_points=100, beta_0=0, beta_1=1.5, sigma=0.10, mu=0)

plt.plot(x, y, 'r')

plt.scatter(x, y)
plt.scatter(a, b)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
line_x = np.linspace(-3, 3, 10)
line_y = np.linspace(-0, 0, 10)

plt.plot(line_x, line_y, 'r')

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data = [31, -41, 59, 26, -53, 58, 97, -93, -23]
cumulative_sum = [sum(data[:i+1]) for i in range(len(data))]

plt.figure(figsize=(10, 7))  # Adjust the figure size if needed
plt.plot(cumulative_sum, marker='o', linestyle='-', color='b')

for i, (cum_sum, change) in enumerate(zip(cumulative_sum, data)):
    plt.annotate(f'{change}', (i, cum_sum), textcoords="offset points", xytext=(0,10), ha='center')

plt.title("the sequence")
plt.xlabel("Index")
plt.ylabel("Cumulative Sum")
plt.grid(True)

plt.show()
```
